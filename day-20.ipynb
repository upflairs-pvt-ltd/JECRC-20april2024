```python
df = pd.read_csv('cleaned_data.csv')
df.head()
```

```python
x = df.drop('Customer Type_disloyal Customer',axis='columns')
y = df[['Customer Type_disloyal Customer']]
```

```python
x.head()
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2)
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
lg = LogisticRegression()
```

```python
import warnings
warnings.filterwarnings('ignore')
```

```python
lg.fit(x_train,y_train)
```

```python
## training set 
round(lg.score(x_train,y_train),2)*100
```

```python
round(lg.score(x_test,y_test),2)*100
```

```python
## OVERFIT PROBLEM 
TRAINING HIGH SCORE -->  98% 
TESTING LOW SCORE   -->  75% 
```

```python
## UNDERFIT PROBLEM 
TRAINING LOW SCORE -->  75% 
TESTING LOW/HIGH SCORE   -->   
```

```python
pred = lg.predict(x_test)
```

```python
y_test['prediction'] = pred 
```
