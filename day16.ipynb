```python
import numpy  as np 
import pandas as pd 
import matplotlib.pyplot as plt 
import seaborn as sns 
```

```python
df = sns.load_dataset('tips')
df.head()
```

```python
df.columns = ['total_bill', 'tip', 'gender', 'smoker', 'day', 'time', 'size']
```

```python
df
```

```python
## relplot()
## catplot()
```

```python
sns.relplot(x='total_bill',y='tip',data=df,hue='gender',style='day',col='day')
```

```python
sns.relplot(x='total_bill',y='tip',data=df,hue='gender',style='day',col='day',kind='line')
```

```python
import warnings
warnings.filterwarnings('ignore')
```

```python
sns.catplot(x='gender',y='total_bill',data=df,kind='violin')
```

```python
sns.scatterplot(x='total_bill',y='tip',data=df)
```

```python
sns.barplot(x='day',y='total_bill',data=df,hue='gender')
plt.title('JECRC')
plt.ylabel("Total Bill")
plt.xlabel("Day")
plt.show()
```

```python
label = list(df['day'].value_counts().keys())
label
```

```python
val = list(df['day'].value_counts().values)
val 
```

```python
sns.catplot(x='smoker',y='total_bill',data=df,kind='bar',hue='day',row='gender') 
```

```python
ls = [0,0,0,0.2]
sns.set(rc={'figure.figsize':(5,5)})
plt.pie(val,labels=label,wedgeprops={'ec':'k'},explode=ls,autopct="%1.2f%%",shadow=True) 
plt.savefig("piechart")
plt.show() 
```
